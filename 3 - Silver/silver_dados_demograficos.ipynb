```python
from pyspark.sql.functions import col
```

```python
%sql
select * from vitivinicultura.bronze_dados_demograficos.dados_demograficos_banco_mundial
```

```python

df = spark.sql("""
    SELECT ROW_NUMBER() OVER (ORDER BY ANO ASC) AS ID_DADO_DEMOGRAFICO, *
    FROM vitivinicultura.bronze_dados_demograficos.dados_demograficos_banco_mundial
    WHERE PAIS IS NOT NULL
""")
df.createOrReplaceTempView("dados_demograficos")
```

```python
 #Cria tabela gerenciada no Unity Catalog
spark.sql("""
CREATE TABLE IF NOT EXISTS vitivinicultura.silver_dados_demograficos.dados_demograficos_banco_mundial
USING DELTA
""")
# Escrita no formato Delta Lake
df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("vitivinicultura.silver_dados_demograficos.dados_demograficos_banco_mundial")

```

```python
from pyspark.sql.functions import col
```

```python
tabela_silver = spark.table("vitivinicultura.silver_dados_demograficos.dados_demograficos_banco_mundial")
```

```python
def gera_tabela_demografica(tabela_silver):
    tabela_silver.createOrReplaceTempView("dados_demograficos")

    df = spark.sql("""
        SELECT 
            ID_DADO_DEMOGRAFICO,
            PAIS,
            ANO,
            CAST(POPULACAO_TOTAL AS BIGINT) AS POPULACAO_TOTAL,
            ROUND(PERCENTUAL_URBANIZACAO, 2) AS PCT_URBANIZADO,
            CAST(PIB_TOTAL AS BIGINT) AS PIB_TOTAL,
            ROUND(PIB_PER_CAPITA, 2) AS PIB_PER_CAPITA,
            ROUND(CRESCIMENTO_PIB, 2) AS CRESCIMENTO_PIB
        FROM 
            dados_demograficos
    """)

    return df
```

```python
# #Cria tabela gerenciada no Unity Catalog
spark.sql("""
CREATE TABLE IF NOT EXISTS vitivinicultura.gold_dados_demograficos.dados_demograficos_banco_mundial
""")
# Escrita no formato Delta Lake
df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("vitivinicultura.gold_dados_demograficos.dados_demograficos_banco_mundial")

```
